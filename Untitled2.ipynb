```python
import clickhouse2pandas as ch2pd
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

connection_url = 'http://thanos:show_me_your_data@dev-01.aksimemor.me:8123'
```

# Dataset queries

```python
revenue_report_query = '''
select registration_period_start, lifecycle_period_start, count(distinct user_id) as users_count, avg(reporting_revenue) as ARPU
from marvel m
group by toStartOfWeek(m.installed_at) as registration_period_start, toStartOfWeek(m.created_at) as lifecycle_period_start
order by registration_period_start, lifecycle_period_start;
'''

activity_report_query = '''
select registration_month, lifecycle_month, user_id
from marvel m
group by toStartOfWeek(m.installed_at) as registration_month, toStartOfWeek(m.created_at) as lifecycle_month, user_id
order by registration_month, lifecycle_month;
'''
```

# Revenue report

## Запрос данных

```python
revenue_report_df = ch2pd.select(connection_url, revenue_report_query)
```

## Формирование пустой матрицы

```python
dates = pd.Index(revenue_report_df.registration_period_start.unique()).union(pd.Index(revenue_report_df.lifecycle_period_start.unique()))
df = pd.DataFrame(columns=dates, index=dates)
```

## Заполнение матрицы

```python
for index, row in revenue_report_df.iterrows():
    df[row.registration_period_start][row.lifecycle_period_start] = row.ARPU
    
df = df.T
df = df.fillna(0)
```

## Вывод матрицы

```python
fig_dims = (21, 14)
fig, ax = plt.subplots(figsize=fig_dims)
ax = sns.heatmap(df, ax=ax)
```

## Выводы

Первые когорты не дают никакого понимания о поведении пользователей, так как включают данные по маленькому числу пользователей. Далее видно, что у пользователей со временем увеличивается ARPU. Возможно, это связано с тем, что в игре со временем остается больше платящих пользователей

# Retention report (not finalized!!!)

```python
activity_df = ch2pd.select(connection_url, activity_report_query)
```

```python
cohort_users = activity_df[['registration_month', 'user_id']]
cohort_users = cohort_users.drop_duplicates()
```

```python
all_users = list(cohort_users['user_id'].unique())
```

```python
def user_retendations(cohort, start_date, user):
    return activity_df[(activity_df['registration_month'] == cohort) & (activity_df['lifecycle_month'] >= start_date) & (activity_df['user_id'] == user)]
```

```python
all_periods = pd.Index(activity_df.registration_month.unique()).union(pd.Index(activity_df.lifecycle_month.unique()))
retention_matrix = pd.DataFrame(columns = all_periods, index=all_periods)
```

```python
for i, row in retention_matrix.iterrows():
    for index, value in retention_matrix[i].items():
        active_users_count = len(list(filter(lambda u: not user_retendations(i, index, u).empty, all_users)))
        print(active_users_count)
        retention_matrix[row][index] = 1
```
